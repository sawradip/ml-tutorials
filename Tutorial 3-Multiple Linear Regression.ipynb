```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
#Importing the dataset


dataset = pd.read_csv('50_Startups.csv')       #dataset is of type DataFrame

x = dataset.iloc[:,:-1].values           #using .values generates array
y = dataset.iloc[:,-1].values             #x & y is of type np.ndarray

print(x)
```

```
#Encoding categorical data

from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder_x = LabelEncoder()                                                 #the categorical data (country & label here) will be  
x[:,3] = labelencoder_x.fit_transform(x[:,3])                                   #transformed into labels(0,1,2.....)
print(x)

columntransformer = ColumnTransformer([('town', OneHotEncoder() , [3])] , remainder = "passthrough")            #but if some countries are labeled with higher values
x = columntransformer.fit_transform(x)                                                                          #that can be misleading, so we transform them to one-hot vectors.
print(x)
```

```
#Avoiding Dummy Variable Trap

x = x[: , 1:]

#print(x)
```

```
#Splitting dataset into train and test set

from sklearn.model_selection import train_test_split

X_train, X_test ,y_train , y_test = train_test_split(x , y ,test_size = .2 , random_state = 0)

print(X_train)
print(X_test)
```

```
#Fitting multiple linear regression to the Training Set

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train , y_train)
```

```
#predicting the Test Set results

y_pred = regressor.predict(X_test)
```

## Backward Elimination method:

```
#Building the optimal model using Backward Elimination method
 
import statsmodels.regression.linear_model as sm                                               
                                                                                #This new library doesn't consider constant b at the model
X = np.append( arr = np.ones((len(x),1)) , values = x , axis = 1)   #Thats why a column of ones is being manually appended with X
                                                                                #axis = 1 is for adding allong the column
X_opt = X[: , [0,1,2,3,4,5]].astype(int)                                        #this type conversion is a must.Otherwise facing the isfinite() input type mismatch problem

regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()          

regressor_OLS.summary()
```

P-value lower than 0.05 means that predictor/feature has significient effect over the predicting value/labels.Otherwise we should ignore the value.

So,in the next step we will consider if the highest p value is over 0.05.If it is,run the model again,without that feature.

```
#Fittting the model again without the highest p value feature(x2)


X_opt = X[: , [0,1,3,4,5]].astype(int)                                        #this type conversion is a must.Otherwise facing the isfinite() input type mismatch problem

regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()          

regressor_OLS.summary()
```

```
#Fittting the model again without the highest p value feature(x1)


X_opt = X[: , [0,3,4,5]].astype(int)                                        #this type conversion is a must.Otherwise facing the isfinite() input type mismatch problem

regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()          

regressor_OLS.summary()
```

```
#Fittting the model again without the highest p value feature(x2)


X_opt = X[: , [0,3,5]].astype(int)                                        #this type conversion is a must.Otherwise facing the isfinite() input type mismatch problem

regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()          

regressor_OLS.summary()
```

```
#Fittting the model again without the highest p value feature(x2)


X_opt = X[: , [0,3]].astype(int)                                        #this type conversion is a must.Otherwise facing the isfinite() input type mismatch problem

regressor_OLS = sm.OLS(endog = y,exog = X_opt).fit()          

regressor_OLS.summary()
```

So,We can safely say that remaining features are very very influential to the predicted value.

This is the Backward Elimination Method.
